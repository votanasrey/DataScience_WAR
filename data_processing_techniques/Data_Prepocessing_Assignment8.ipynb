<h1>Apply the Text Preprocessing using TF and IDF<h1>

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()


text = """
The price of the shirt is $19.99, while the shoes cost $59.95.
We also have a special offer: buy two shirts for $30.
In addition, there is a sale on accessories: earrings for $9.99 and bracelets for $14.50.
Don't miss out on these amazing deals!
"""

# Define a regular expression pattern to match prices
price_pattern = r'\$\d+\.\d{2}'

# Find all matches of prices in the text using regular expressions
prices = re.findall(price_pattern, text)

# Join the matched prices into a single string
cleaned_text = ' '.join(prices)

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the cleaned text using TF-IDF vectorizer
tfidf_matrix = tfidf_vectorizer.fit_transform([cleaned_text])

# Get the feature names (terms) from the TF-IDF vectorizer
feature_names = tfidf_vectorizer.get_feature_names_out()


print("Original text:")
print(text)
print("\nExtracted prices:")
print(prices)
print("\nCleaned text:")
print(cleaned_text)
```
